### Till att börja med har jag tänkt att kolla på och utforska följande:

Hur den dagliga ökningen av nya fall såg ut i Sverige.

```python
# Daglig ökning av Covid-19 fall i Sverige
fig_daglig_ökning = px.area(covid19_data,
                            x='Statistikdatum',
                            y='Totalt_antal_fall',
                            title='Daglig ökning av Covid-19-fall i Sverige',
                            labels={'Daglig_ökning': 'Antal fall', 'Statistikdatum': 'Datum'})

fig_daglig_ökning.show()


# Spara till HTML
fig_daglig_ökning.write_html('../Labb/Visualiseringar/Uppgift_3/fig_daglig_ökning.html')
```

## **Daglig ökning av Covid-19-relaterade dödsfall:** 

Närmare titt på antalet dagliga dödsfall.

```python
# **Daglig ökning av Covid-19-relaterade dödsfall:**

dödsfall_flik = 'Antal avlidna per dag'
covid_dödsfall_data = pd.read_excel(filväg_covid19, sheet_name=dödsfall_flik)
covid_dödsfall_data.head()
```

```python
fig_dödsfall = px.line(covid_dödsfall_data,
                       x='Datum_avliden',
                       y='Antal_avlidna',
                       title='Daglig ökning av Covid-19 dödsfall i Sverige',
                       labels={'Datum_avliden': 'Datum', 'Antal_avlidna': 'Antal avlidna'})

fig_dödsfall.show()
fig_dödsfall.write_html("../Labb/Visualiseringar/Uppgift_3/fig_dödsfall.html")
```

## **Intensivvård, fall & avlidna för könen:** 

Jag vill titta på totala antalet Covid19-fall samt hur många som blivit inlagda på intensivvård och även titta på deras fördelning för könen.

```python
kön_flik = 'Totalt antal per kön'
covid_kön_data = pd.read_excel(filväg_covid19, kön_flik)
display(covid_kön_data.head())

# Jag gör ett antagande att Uppgift saknas i kolumnen för kön refererar till fall där kön inte definieras som varken man eller kvinna
# väljer att döpa om den till något mer lämpligt.
covid_kön_data.loc[covid_kön_data['Kön'] == 'Uppgift saknas', 'Kön'] = 'Ej specifierat'

fig_kön_data = px.bar(covid_kön_data,
                      x='Kön',
                      y=['Totalt_antal_fall', 'Totalt_antal_intensivvårdade', 'Totalt_antal_avlidna'],
                      title='Antal fall, intensivvårdade & avlidna per kön',
                      labels={'value': 'Antal', 'variable': 'Kategori'},
                      log_y=True,
                      barmode='group'
)
fig_kön_data.show()
fig_kön_data.write_html("../Labb/Visualiseringar/Uppgift_3/fig_kön_data.html")
```

## Effekten av Covid19 baserat på ålder:

Liknande ovan diagram fast här vill jag titta närmre på åldersgrupper istället för könsuppdelning.

```python
flik_blad = "Totalt antal per åldersgrupp"

df = pd.read_excel(filväg_covid19, sheet_name= flik_blad)

# döper om åldersgrupp, tar bort "Ålder_", byter "_plus" till "+", samt ersätter "_" med "-" 
df["Åldersgrupp"] = df["Åldersgrupp"].replace({"Ålder_": "", "_plus": "+", "_": "-"},regex=True)
df

# Uppgift saknas även i detta datasetet. Här står den här raden för en sån liten andel av datat i stort att det inte 
# märkbart kommer påverka om jag tar bort raden.

df.drop(df.loc[df["Åldersgrupp"] == "Uppgift saknas"].index, inplace = True)
df
```

Plottar antal fall, intensivvårdade samt dödsfall för olika åldersgrupper i Sverige:

```python
fig_åldersgrupper = px.bar(
    df,
    x="Åldersgrupp",
    y=["Totalt_antal_fall", "Totalt_antal_intensivvårdade", "Totalt_antal_avlidna"],
    barmode="group",
    log_y=True,
    title="Påverkan på olika åldersgrupper i Sverige",
    labels= {"value": "Antal människor", "variable": ""}
)
fig_åldersgrupper
fig_åldersgrupper.write_html("../Labb/Visualiseringar/Uppgift_3/fig_påvärkan_åldersgrupper.html")
```
